```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100)*10
```

```python
y = 2.3 * x + 100 + np.random.uniform(low=-5, high=5, size=(100))
```

```python
plt.scatter(x,y)
```

```python
def compute_cost(x, y, w, b):

    m = x.shape[0]
    cost = 1 / (2 * m) * np.dot(w * x + b - y, w * x + b - y)
    return cost
```

```python
w_init = 0
b_init = 0
compute_cost(x, y, w_init, b_init)
```

```python
def compute_gradients(x, y, w, b):

    m = x.shape[0]
    dJ_dw = 0
    dJ_db = 0
    for i in range(m):
        dJ_dw += (w*x[i] + b - y[i])*x[i]
        dJ_db += (w*x[i] + b - y[i])
    dJ_dw /= m
    dJ_db /= m
    return dJ_dw, dJ_db
```

```python
def gradient_descent(x, y, w_init, b_init, alpha=0.01, n_iters=10000):

    m = x.shape[0]
    w = w_init
    b = b_init

    for i in range(n_iters):
        dJ_dw, dJ_db = compute_gradients(x, y, w, b)
        w = w - alpha * dJ_dw
        b = b - alpha * dJ_db

        if i % (n_iters / 50) == 0:
            cost = compute_cost(x, y, w, b)
            print(f"{cost=}, {w=}, {b=}")

    return w, b
```

```python
w, b = gradient_descent(x, y, w_init=0, b_init=0, alpha=0.01, n_iters=10000)
```

```python
plt.scatter(x,y)
plt.plot(x,w*x+b,'g')
```
